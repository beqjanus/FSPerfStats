# Firestorm usage analysis Feb 2022 vs Feb 2021 <a class=tocSkip>
# Introduction
This notebook is a work in progress analysing statistical data provided by LL.
The data we have are based on session information with two month's of data, spaced 12 months apart; February 2021 and February 2022
Using this data we will attempt to answer a few important questions

    1) What proportion of our user base are restricted to  32-bit viewer builds by either hardware or operating system?
    2) What is the hardware demographic of our user base and can we discern a rate at which the user base upgrades?
    3) What is the Operating System demographic, how concerned are we by legacy OS support, versus bleeding edge support. Notably Windows 7 versus Windows 11?
    4) How many users are out of reach of our support, Apple Silicon, for which we have little or no capability versus low-end notebook computers?
    5) What proportion of the user base is using a laptop?
    6) What proportion of the laptop user base does not have a discrete graphics card?
    7) Can we gauge the impact of a move to Vulkan by estimating the current and future exposure?

## Data quality issues/concerns
### Firestorm session data bias
The raw data itself is based on around 8 million sessions each month logged. Session based data is, however, potentially biased. If a session is recorded with each discrete login then there will be an innate bias in the figures favouring all configurations that restart more frequently. As such we might see a slight over stating of weak platforms that are more inclined to crash, for example. The biggest single bias would be in terms of laptops, a laptop user is far more likely to logout whenever they leave the room or go back to RL for any extended period. Desktop users on the other hand will have a tendency to stay logged in for extended periods.

### Reference data
To examine the hardware age I have had to source reference data from the primary CPU makers.
Intel - The dataset was scraped directly from https://ark.intel.com using python/scrapy, as such I can regenerate an updated version when required. The dataset is large but required extensive cleaning.
AMD - A dataset was obtained from Kaggle.com. The dataset is quite limited, with CPUs only having an identified launch date going back to 2016 and is not maintained going forward. A more complete and sustainable source needs to be identified.
Apple - Apple silicon (for MacOS) is new enough that I could hand code the data from reference sources.

## AMD data
AMD data is poor quality, a better reference is needed. The data we have does appear to cover a significant proportion of our AMD-based users.

```python

def load_reference_data(name, index='id', date_from_column='launch_date'):
    ref_dir = "./reference/"

    return pd.read_csv(
        ref_dir+name+".csv",
        index_col=index,
        parse_dates={"date":[date_from_column]})

amd_processors = load_reference_data("amd_processors")

amd_processors['processor_number'] = amd_processors['name'].str.lower().str.replace("\s","",regex=True)
amd_processors.loc[~(amd_processors['date'].isna()),'64'] = True  # Every AMD chip that has a valid date is 64-bit (AMD stopped making 32-bit CPUS around 2005)
amd_processors.loc[(amd_processors['64'].isna()),'64'] = True  # For now set everything else 64
amd_processors.loc[amd_processors['vertical_segment']=='Laptop', 'vertical_segment'] = 'Mobile'
display(amd_processors.head())
```

## Cleanup the data sourced from Intel

* Normalise processor_number to lowercase

### We need to know if chips are 32 bit or not. 
The '64' column is not fully populated from ark so we need to clean that. 
We attempt to compute this based, the instruction set, the processor name (all Itaniums are 64-bit), and some use of timeline knowedge (Celerons post 2005 are 64-bit)

## Assign the CPU Make where possible
The CPU Make is determined by a relatively simple set of regular expressions.
The Regex are not optimised in order to make it simpler to add new groups as required.

There are 5 categories at present, AMD, Intel, Apple, Virtual and PS4

Intel/AMD applies to all Intel sessions irrespective of OS. 

Virtual is for the most part unclassifiable and will be ignored as appropriate. 

PS4 is actually two AMD chips that were only ever sold through Sony and are now (apparently) running Firestorm Linux

```python
def extract_CPU_make(df):
    df['cpu_make'] =  df['cpu'].str.extract('^\s*(Intel|AMD|Apple)')
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('Pentium')), 'cpu_make'] =  'Intel' # for any pentiums that are not already assigned set them to Intel
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('Celeron')), 'cpu_make'] =  'Intel' # same for Celeron
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('Radeon')), 'cpu_make'] =  'AMD' # AMD specifics
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('A4-6210')), 'cpu_make'] =  'AMD' # ditto
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('\+8G')), 'cpu_make'] =  'AMD' # best guess corrupted entry likely extreme edition Radeon +8G
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('Quad-Core Processor')), 'cpu_make'] =  'AMD' # best guess corrupted entry likely Phenom X4 range
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('Virtual')), 'cpu_make'] =  'Virtual' # Lots of virtuals which don't help us much
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('Common KVM processor')), 'cpu_make'] =  'Virtual' # Lots of virtuals which don't help us much
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('[01][0-9]/[0-9](?:[a-z])?')), 'cpu_make'] =  'Virtual' # Lots of virtuals which don't help us much
    df.loc[(df['cpu_make'].isna()) & (df['cpu'].str.contains('DG1002FGF84HT|DG1101SKF84HV')), 'cpu_make'] =  'PS4' # Yep really!
    return df['cpu_make']

fs_cpu_gpu_feb_22['cpu_make'] = extract_CPU_make(fs_cpu_gpu_feb_22)
fs_cpu_gpu_feb_21['cpu_make'] = extract_CPU_make(fs_cpu_gpu_feb_21)

# display(fs_cpu_gpu_feb_22['cpu_make'].value_counts(dropna=False))
# display(fs_cpu_gpu_feb_21['cpu_make'].value_counts(dropna=False))
# display(fs_cpu_gpu_feb_22[fs_cpu_gpu_feb_22['cpu_make'].isna()])
# display(fs_cpu_gpu_feb_21[fs_cpu_gpu_feb_21['cpu_make'].isna()])
total_sessions_22 = fs_cpu_gpu_feb_22['CNT'].agg('sum')
total_sessions_21 = fs_cpu_gpu_feb_21['CNT'].agg('sum')

display(Markdown("### Total sessions Feb 2021: " + str(total_sessions_21)))
display(Markdown("### Total sessions Feb 2022: " + str(total_sessions_22)))

def plot_vendor_pie(df):
    # display(df[['cpu_make','count']].groupby('cpu_make')['count'].agg('sum'))
    num_vendors = df[['cpu_make','CNT']].groupby('cpu_make')['CNT'].agg('sum').agg('count')
    explode = (0.1,) * num_vendors
    df[['cpu_make','CNT']].groupby('cpu_make')['CNT'].agg('sum').plot.pie(y='sum',autopct='%1.1f%%', shadow=False, explode = explode, startangle=45,labeldistance = 1.2)
    # ax = ((df.groupby('cpu_make')['count'].agg(['sum']) / total_sessions)*100).plot.pie(y='sum')
    ax.set_title("% Sessions by CPU Vendor")
    ax.set_xlabel("% Sessions")
    ax.set_ylabel("")
    # plt.show()
    # ax.set_xticklabels(x_labels)

```

## Cleaning up the Session data more.

### The CPU data is quite dirty, lets have a quick stab at cleaning it up
The CPU data captured by the viewer is very low quality though this is mostly down to inconsistent labelling by the CPU manufacturers

```python
def cleanup(df):
    df['cpu'] = df['cpu'].str.replace('( CPU )?@.*$','',regex=True) # trim the trailing junk that we're not using right now
    df['cpu'] = df['cpu'].str.replace('( CPU )',' ',regex=True) # remove the word CPU
    df['cpu'] = df['cpu'].str.replace('(\(R\))','',regex=True) # remove the '(R)'
    df['cpu'] = df['cpu'].str.replace('(\(TM\))','',regex=True) # remove the '(TM)'
    df['cpu'] = df['cpu'].str.replace('(i[3579]) +([0-9]+)',r'\g<1>-\g<2>',regex=True) # replace i7 960 with i7-960
    df['cpu'] = df['cpu'].str.replace('( \([0-9.]+ ?MHz\){1}.*)$','',regex=True) # trim the trailing junk that we're not using right now
    # df['cpu'] = df['cpu'].str.replace('(Xeon.*E[0-9][^0-9]+[0-9]{4})( v)([0-9])',r'\g<1>V\g<3>',regex=True) # replace lowercase v in Xeons andclean whitespace
    df['cpu'] = df['cpu'].str.replace('(Xeon.*E5\-267)8',r'\g<1>0',regex=True) # E5-2678 is not in ARK 2670 is and has same date :-(
    df['cpu'] = df['cpu'].str.replace('(Xeon.*W\-21[45])\d',r'\g<1>5',regex=True) # 
    df['cpu'] = df['cpu'].str.replace('(i[357]\-10[0-9])[0-9]N(\S*)',r'\g<1>0\g<2>',regex=True) # remove the N from 10th gen and clamp to power class 0
    df['cpu'] = df['cpu'].str.replace('(i9\-109)[0-9]{2}(\S*)(?<!(XE|HK))',r'\g<1>00\g<2>',regex=True) # 10th gen i9 hpx clamp to power class 0
    df['cpu'] = df['cpu'].str.replace('(i9\-109)[0-9]{2}(\S*)(?<=(XE|HK))',r'\g<1>80\g<2>',regex=True) # except XE and HK clamp to power class 8
    df['cpu'] = df['cpu'].str.replace('(i9\-108)[0-9]{2}(\S*)(?<!(H))',r'\g<1>50\g<2>',regex=True) # 10th gen i9 hp clamp to power class 5
    df['cpu'] = df['cpu'].str.replace('(i9\-108)[0-9]{2}(\S*)(?<=(H))',r'\g<1>85\g<2>',regex=True) # except H which is 8
    # df['cpu'] = df['cpu'].str.replace('((Core) m([0-9]\-[0-9]{1}Y[0-9]{2})',r'\g<1> M\g<2>',regex=True) # core m needs M
    # df['cpu'] = df['cpu'].str.replace('(Xeon.*E[0-9][^0-9]+[0-9]{2})([0-9]{2})( v)([0-9])',r'\g<1>00V\4',regex=True) 
    df['cpu'] = df['cpu'].str.replace(r"\s+"," ", regex=True) # replace multi white space with single space
    df['cpu'] = df['cpu'].str.replace(r"(^\s+)|(\s+$)","", regex=True) # clear leading/trailing space
    return df

fs_cpu_gpu_feb_22 = cleanup(fs_cpu_gpu_feb_22)
fs_cpu_gpu_feb_21 = cleanup(fs_cpu_gpu_feb_21)
```

## Extract the cpu code to allow lookup against the vendor ref data
There's a few guesses made here. However, failure to produce a valid code will fail the launch date lookup later. Processor codes are downcased for matching.

```python
# fs_cpu_gpu_feb_22['cpu_code']
def create_processor_number(row):
    intel_cpu_code_re = [
        r'((W\-*([0-9]+)?))',
        r'(E[0-9]\- *[0-9Mm]+\s*([vV][0-9]+)?)', # patterns matching Wnnnn or En-nnnnVnn
        r'([NGQJEPT]\b[0-9]{4}\b)', # patterns match a prefix letter and 4 exactly digits
        r'([0-9]{4}[UM])', # ddddU
        r'(i[0-9]-[0-9]+[\S]+)', # typical modern core i series. forms such as i7-1337OMG 
    ]
    # AMD is not well developed yet
    amd_cpu_code_re = [
        r'(AMD Ryzen [0-9] [0-9]+[\S])',
        r'((AMD Athlon)(?:[\S]+)? ?(?:[\S]+ ?){0,2} ?[0-9]{3,4}([\S])?)' ]
    apple_cpu_code_re = [
        r'(Apple M[0-9]( (Max|Pro|Ultra))?)'
    ]
    try:
        # display(row.head())
        cpu = row["cpu"]
        if row['cpu_make'] == 'Intel':
            # row['processor_number'] = row['cpu'].str.extract('()', expand=True)
            grps = re.search('(' + '|'.join(intel_cpu_code_re) + ')', cpu)
            if grps is not None:
                return grps[1]
            # Some need a remapping in order to find a match
            grps = re.search('(i[3579]) ?([QUKMTX]*) (\S+)$', cpu) # i7 Q 4578 -> i7-4578Q 
            if grps is not None:
                if grps[2] == 'Q' or grps[2] == 'U':
                    return grps[1]+'-'+grps[3]+grps[2]+'M'
                else:
                    return grps[1]+'-'+grps[3]+grps[2]
            grps = re.search('Xeon (E[0-9])([0-9]{4})', cpu) # E31245
            if grps is not None:
                return grps[1]+'-'+grps[2]
            grps = re.search('Atom (\S+)$', cpu) # Atoms just need trailing end. pick that up explicitly
            if grps is not None:
                return grps[1]
            grps = re.search('Core M-(\S+)$', cpu) # Atoms just need trailing end. pick that up explicitly
            if grps is not None:
                return grps[1]
            grps = re.search(r'(\S+)$',cpu)
            if grps is not None:
                # print("***" + cpu+'('+grps[1]+')')
                return grps[1]
            return None

        if row['cpu_make'] == 'Apple':
            # row['processor_number'] = row['cpu'].str.extract('()', expand=True)
            try:
                return re.search('(' + '|'.join(apple_cpu_code_re) + ')', cpu)[0]
            except TypeError:
                    return None
        if row['cpu_make'] == 'AMD':
            # row['processor_number'] = row['cpu'].str.extract('()', expand=True)
            try:
                return re.search('(' + '|'.join(amd_cpu_code_re) + ')', cpu)[0]
            except TypeError:
                    return None
    except KeyError:
        return None

fs_cpu_gpu_feb_22['processor_number'] = fs_cpu_gpu_feb_22.apply(create_processor_number,axis=1)
fs_cpu_gpu_feb_21['processor_number'] = fs_cpu_gpu_feb_21.apply(create_processor_number,axis=1)
fs_cpu_gpu_feb_22['processor_number'] = fs_cpu_gpu_feb_22['processor_number'].str.lower().str.replace("\s","",regex=True)
fs_cpu_gpu_feb_21['processor_number'] = fs_cpu_gpu_feb_21['processor_number'].str.lower().str.replace("\s","",regex=True)
```

```python
def combine_2y_agg(year1, year2, y1label="Y1", y2label="Y2"):
    y1 = get_count_by_cpu_age(year1)
    y2 = get_count_by_cpu_age(year2)
    combined = pd.concat([y2.rename(y2label), y1.rename(y1label)], axis=1)
    tot2 = combined[y2label].agg('sum')
    tot1 = combined[y1label].agg('sum')

    combined[y2label+'pct'] = combined[y2label]/tot2*100
    combined[y1label+'pct'] = combined[y1label]/tot1*100
    combined['changepct'] = combined[y2label+'pct']-combined[y1label+'pct']
    combined['change'] = combined[y2label]-combined[y1label]
    combined['rateofpctchange'] = combined['changepct']/combined[y1label+'pct']*100
    return combined
combined = combine_2y_agg(all_with_extras_21,all_with_extras_22,y1label="Feb21", y2label="Feb22")
display(combined)
```

```python
from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import PercentFormatter
from matplotlib.ticker import LogFormatter

class CustomLog(LogFormatter):
    def __call__(self, val, pos = None):
        if val<1000:
            return ScalarFormatter.__call__(self, val) 
        else:
            return LogFormatter.__call__(self, val)

class CustomLogPercent(LogFormatter,PercentFormatter):
    def __init__(self):
        LogFormatter.__init__(self)
        PercentFormatter.__init__(self)
                
    def __call__(self, val, pos = None, xmax=100):
        if abs(val)<10000:
            return PercentFormatter.__call__(self, val) 
        else:
            return LogFormatter.__call__(self, val)+"%"

def plot_multiyear_cpu_age(timeline, title=None, logscale=False, ylabel = None, percent=True, ax=None):
    # if not percent and not logscale:
    #     plt.ticklabel_format(style='plain', axis='y')
    ax = timeline.plot.bar(ax=ax)
    plt.subplots_adjust(bottom=.25, left=.25)
    ax.xaxis.set_major_formatter(
        FuncFormatter(lambda x, p: str(timeline.index[x])[0:4] if str(timeline.index[x])[0] != '<' else str(timeline.index[x])))
    if logscale:
        ax.set_yscale('symlog')
        if percent:
            ax.yaxis.set_major_formatter(CustomLogPercent())
        else:
            ax.yaxis.set_major_formatter(CustomLog())
        if ylabel is not None:
            ylabel =ylabel + '\n(log scale)'
    else:
        if percent:
            ax.yaxis.set_major_formatter(PercentFormatter())
        else:
            ax.yaxis.set_major_formatter(
                FuncFormatter(lambda x, p: format(int(x), ',')))
    if title is not None:
        ax.set_title(title)
    if ylabel is not None:
        ax.set_ylabel(ylabel)
    return ax
    
```

```python
# merge_older = 
def summarise_older(df, tot1, tot2, y1label='Y1', y2label='Y2'):
        summarised = df.agg('sum')
        if tot2 !=0:
                summarised[y2label + 'pct'] = summarised[y2label]/tot2*100
        if tot1 !=0:
                summarised[y1label + 'pct'] = summarised[y1label]/tot1*100
        summarised['changepct'] = summarised[y2label + 'pct']-summarised[y1label + 'pct']
        summarised['change'] = summarised[y2label]-summarised[y1label]
        summarised['rateofpctchange'] = summarised['changepct']/summarised[y1label + 'pct']*100
        return summarised

combined_clean = combined.copy()
# timeline.drop(labels=['2010','2009','2008','2007','2006','2005','2004'], inplace=True)
```

```python
def summarise_and_clean(df, y1label='Y1', y2label='Y2',fromYear=None, toYear='2011-1-1',):
    tot2 = df[y2label].agg('sum')
    tot1 = df[y1label].agg('sum')
    older_machines = df.loc[fromYear:toYear]
    summarised=summarise_older(older_machines, tot1, tot2, y1label=y1label, y2label=y2label)
    clean = df.drop(older_machines.index)
    summarised.rename("<2011", inplace=True)
    sumdf = pd.DataFrame([summarised])
    clean = pd.concat([sumdf,clean])
    clean[y2label + 'cumpct'] = clean[[y2label + 'pct']].cumsum(0)
    clean[y1label + 'cumpct'] = clean[[y1label + 'pct']].cumsum(0)
    return clean

combined_clean = summarise_and_clean(combined, y1label='Feb21', y2label='Feb22')
```

## GPUs - What can we learn from GPUs
### Data Prep: cleanup and enrich
We have no reference data for GPU at the moment, so we do some basic cleanup to assign a make and tease out some curios.

In future, it would be interesting to identify feature support.

```python
def extract_GPU_make(df):
    df['gpu'] =  df['gpu'].str.lower()
    df['gpu_make'] = df['gpu'].str.extract('(nvidia|apple|amd)\s')
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('mesa')), 'gpu_make'] =  'software' # software on Linux 
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('nouveau')), 'gpu_make'] =  'nvidia' # nvidia on Linux with open source drivers
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('geforce')), 'gpu_make'] =  'nvidia' # nvidia 
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('radeon')), 'gpu_make'] =  'amd' # amd
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('ati\s')), 'gpu_make'] =  'amd' # amd
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('llvmpipe')), 'gpu_make'] =  'software' # llvm gallium softwar drivers
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('d3d12')), 'gpu_make'] =  'software' # MS d3d12 debug layer driver
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('virgl')), 'gpu_make'] =  'virtual' # virtual on Linux 
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('vmware')), 'gpu_make'] =  'virtual' # vmware virtual (may be software)
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('amazon web services')), 'gpu_make'] =  'virtual' # virtual on Linux 
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('imagination technologies')), 'gpu_make'] =  'other' # virtual on Linux 
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('broadcom')), 'gpu_make'] =  'other' # virtual on Linux 
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('zink')), 'gpu_make'] =  'software' # collarbora zink opengl on vulkan
    df.loc[(df['gpu_make'].isna()) & (df['gpu'].str.contains('intel')), 'gpu_make'] =  'intel' # this has to go last as others include intel too
    return df['gpu_make']

all_with_extras_21['gpu_make'] = extract_GPU_make(all_with_extras_21)
all_with_extras_22['gpu_make'] = extract_GPU_make(all_with_extras_22)
```

# Summary of results
We set out to answer a few questions. The answers need to be caveated with a reminder that this is session-based data and as such may contribute a bias that we have no way to remove. Hopefully future, unbiased data will allow us to re-run this analysis and draw more compelling conclusions.

    1) What proportion of our user base are restricted to  32-bit viewer builds by either hardware or operating system?
    We have shown that the number of 32-bit viewer sessions (as shown in the separate usage stats that we get periodically) are not for the most part driven by necessity, at least from an OS or CPU perspective with a far smaller proportion of sessions coming from those legacy platforms than would otherwise be the case. One, frequently cited, reason that users run the 32-bit viewer is memory use. The 4GB limit providing a form of containment. Others claim it to be faster. Further analysis could be conducted to establish how many 32-bit users are running on machines with limited RAM. 

    2) What is the hardware demographic of our user base and can we discern a rate at which the user base upgrades?
    With just two samples it is not possible to confidently predict trends but the data suggests that the population overall stretches back up to 15 years with the 95th percentile at around 10 years and the median at 4 years. This implies a relatively even distribution over the period. The falloff over time is surprisingly constant (around 15%-20% year on year in any given age group beyond 4 years). When users upgrade the majority purchase the latest generation of machine, with about 1/3 of users opting for the better value option of the last year's machine. 

    3) What is the Operating System demographic, how concerned are we by legacy OS support, versus bleeding edge support. Notably Windows 7 versus Windows 11?
    Our users are for the most part on the latest stable release across all platforms, with Win 11 and MacOs Monterey starting to grow. There is also a hard core of people who are change averse which aligns with our expectations based on when users update their viewer. Undoubtedly a proportion of those users are waiting for ALL the bugs to be fixed in Windows 10 before they'll move.
    4) How many users are out of reach of our support, Apple Silicon, for which we have little or no capability versus low-end notebook computers?
    The sample data shows only a tiny proportion of users on Apple Silicon but that grew at a high rate between the two sample points. More data is required to get a conclusive view.
    We have not analysed based on performance yet, and thus have no indication of "low power performance". However, we can see from the data that there are users running hardware that is barely within the woefully out of date LL guidelines (themselves over 12 years since last update IIRC).
    5) What proportion of the user base is using a laptop?
    A little less than half the user base is using hardware powered by a CPU designated for mobile device use (i.e. laptops). 
    6) What proportion of the laptop user base does not have a discrete graphics card?
    Only around 20% of laptop users in the data sets are using the onboard graphics. Caveat: This analysis was conducted for Intel mobile CPUs only.
    7) Can we gauge the impact of a move to Vulkan by estimating the current and future exposure?
    Vulkan adoption is limited my a number of factors but the hardest barrier is for Intel Laptops on Integrated graphics, that were made before 2017. less than half and falling quite rapidly.
